## Task
Modify the dataframe `df` to have 100 rows. Add a column named 'IMC' that calculates the Body Mass Index and a column named 'clasificacion' that classifies the individual based on their IMC (underweight, normal weight, overweight, etc.).

## Generate data

**Reasoning**:
Generate additional random data for name, weight, and height to reach 100 rows and concatenate it with the existing dataframe.



```python
import numpy as np

additional_rows_needed = 100 - len(df1)

# Generate random data for the new rows
random_names = [f"Person_{i}" for i in range(additional_rows_needed)]
random_weights = np.random.uniform(40, 100, additional_rows_needed) # realistic weight range
random_heights = np.random.uniform(1.50, 1.90, additional_rows_needed) # realistic height range

# Create a new DataFrame with the random data
new_data = {'nombre': random_names, 'peso': random_weights, 'altura': random_heights}
new_df = pd.DataFrame(new_data)

# Concatenate the original DataFrame with the new DataFrame
df1 = pd.concat([df1, new_df], ignore_index=True)

# Display the updated dataframe to verify the number of rows
display(df1)
```

## Calculate imc

### Subtask:
Add a new column named 'IMC' to the dataframe and calculate the Body Mass Index for each row using the 'peso' and 'altura' columns. The formula for BMI is weight (kg) / height (m)^2.


**Reasoning**:
Add a new column 'IMC' to the dataframe and calculate the BMI using the 'peso' and 'altura' columns.



```python
df1['IMC'] = df1['peso'] / (df1['altura'] ** 2)
display(df1.head())
```

## Classify based on imc

### Subtask:
Add a new column named 'clasificacion' and classify each individual based on their calculated IMC. The classifications will be based on standard BMI categories (e.g., Underweight, Normal weight, Overweight, Obesity).


**Reasoning**:
Define a function to classify IMC and apply it to the 'IMC' column to create the 'clasificacion' column.



```python
def classify_imc(imc):
    if imc < 18.5:
        return 'Underweight'
    elif 18.5 <= imc < 25:
        return 'Normal weight'
    elif 25 <= imc < 30:
        return 'Overweight'
    else:
        return 'Obesity'

df1['clasificacion'] = df1['IMC'].apply(classify_imc)
display(df1.head())
```

#EDA (ANLISIS DE DATOS EXPLORATORIO)

**Anlisis de Datos Exploratorio (EDA, Exploratory Data Analysis)** es el **primer paso fundamental** cuando trabajas con cualquier dataset, ayuda a **conocer los datos, limpiarlos y entender sus patrones** antes de aplicar modelos ms avanzados.

**PASOS**

1. **Cargar y revisar el dataset**

* Importar libreras (`pandas`, `numpy`, `matplotlib`, `seaborn`).
* Cargar el dataset con `pd.read_csv()` u otro mtodo.
* Revisar las primeras filas con `df.head()`.
* Ver el tamao de los datos: `df.shape`.

2. **Inspeccin general**

* Ver los **tipos de datos**: `df.info()`.
* Identificar **valores nulos o faltantes**: `df.isnull().sum()`.
* Revisar valores duplicados: `df.duplicated().sum()`.

3. **Estadstica descriptiva**

* Para **variables numricas**: `df.describe()`.
* Para **variables categricas**: `df.describe(include="object")` o `value_counts()`.
* Medidas importantes: media, mediana, moda, rango, varianza, desviacin estndar.

4. **Anlisis univariado (una variable a la vez)**

* **Numricas**  histogramas, boxplots (para ver distribucin y outliers).
* **Categricas**  tablas de frecuencia, grficos de barras.

5. **Anlisis bivariado (relaciones entre dos variables)**

* Correlaciones entre numricas: `df.corr()`.
* Diagramas de dispersin (`scatterplot`) para relaciones.
* Tablas cruzadas (`pd.crosstab`) para categricas.
* Grficos combinados (ej: boxplot de variable numrica segn categora).

6. **Anlisis multivariado**

* Mapas de calor de correlaciones (`heatmap`).
* Anlisis de distribucin conjunta (pairplot en `seaborn`).
* Clustering visual bsico (si aplica).

7. **Deteccin de outliers y valores atpicos**

* Usar **boxplots** para detectar datos extremos.
* Comparar rangos intercuartlicos (IQR).

8. **Limpieza y preparacin de los datos**

* Tratar valores nulos (eliminarlos o imputarlos).
* Normalizar o estandarizar variables numricas.
* Codificar variables categricas (OneHotEncoder, LabelEncoder).
* Eliminar duplicados o inconsistencias.

9. **Visualizacin y conclusiones**

* Graficar patrones clave (tendencias, distribuciones, correlaciones).
* Resumir hallazgos en un **informe corto**:

  * Qu variables explican mejor el dataset?
  * Qu problemas de calidad de datos existen?
  * Qu hiptesis iniciales se pueden plantear?

 **En resumen**:
El EDA es un proceso iterativo que combina **estadstica descriptiva + visualizacin de datos** para conocer la estructura y patrones antes de aplicar modelos de machine learning o anlisis avanzados.

#REGRESIN LINEAL

Cuando trabajas con un modelo de regresin, necesitas mtricas que te permitan medir **qu tan bien predice los valores continuos en tu dataset**.

 **Mtricas de Regresin**
1. **Error Absoluto Medio (MAE  Mean Absolute Error)**

* Promedio de las diferencias absolutas entre valores reales y predichos.
* Fcil de interpretar (mide en las mismas unidades de la variable).
* Cuanto ms pequeo, mejor.

2. **Error Cuadrtico Medio (MSE  Mean Squared Error)**

* Penaliza ms los errores grandes porque eleva al cuadrado.
* Se usa mucho en optimizacin de modelos.
* Entre ms pequeo, mejor.

3. **Raz del Error Cuadrtico Medio (RMSE  Root Mean Squared Error)**

* Es la raz del MSE.
* Muy til porque vuelve al mismo rango de la variable dependiente.
* Ms interpretable que el MSE.

4. **Error Absoluto Medio Porcentual (MAPE  Mean Absolute Percentage Error)**

* Expresa el error en porcentaje.
* Intuitivo, pero puede ser problemtico si hay valores reales cercanos a 0.

5. **Coeficiente de Determinacin (R)**

* Indica la proporcin de la variabilidad de la variable dependiente explicada por el modelo.
* Valores cercanos a 1  buen ajuste.
* Valores negativos  el modelo es peor que predecir con la media.

6. **R Ajustado (Adjusted R)**

* Variante del R que ajusta segn el nmero de variables en el modelo.
* Evita que aadir variables irrelevantes aumente artificialmente el R.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Supongamos que tienes:
y_real = [3, 5, 7, 10]
y_pred = [2.5, 5.1, 6.8, 9.5]

# MAE
mae = mean_absolute_error(y_real, y_pred)

# MSE
mse = mean_squared_error(y_real, y_pred)

# RMSE
rmse = np.sqrt(mse)

# R
r2 = r2_score(y_real, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R:", r2)
```

##MATRIZ DE CORRELACIN

Una matriz de correlacin es una herramienta estadstica que muestra **cmo se relacionan (linealmente) entre s las variables numricas de un dataset.**

 Qu es la correlacin?

La correlacin mide el grado de asociacin lineal entre dos variables.
Su valor est entre -1 y 1:

+1  correlacin positiva perfecta (si una sube, la otra tambin).
-1  correlacin negativa perfecta (si una sube, la otra baja).
0  no hay correlacin lineal.

 Qu es la matriz de correlacin?

Es una tabla cuadrada que contiene todos los **coeficientes de correlacin entre pares de variables numricas.**
La diagonal principal siempre es 1, porque cada variable est perfectamente correlacionada consigo misma.

Se usa para:
* Detectar relaciones entre variables.
* Ver redundancia (variables muy correlacionadas).
* Prevenir multicolinealidad en modelos de regresin.

#MATRIZ DE CONFUSIN

Es una tabla de doble entrada que se utiliza para **evaluar el rendimiento de un modelo de clasificacin.**

Muestra cuntas predicciones fueron correctas y equivocadas, organizadas por clases.

 Ejemplo de una matriz de confusin binaria (dos clases: Positivo y Negativo):
|                   | Predicho Positivo       | Predicho Negativo       |
| ----------------- | ----------------------- | ----------------------- |
| **Real Positivo** | TP (Verdadero Positivo) | FN (Falso Negativo)     |
| **Real Negativo** | FP (Falso Positivo)     | TN (Verdadero Negativo) |

 Significado:

* TP (True Positive)  Casos positivos correctamente predichos.
* TN (True Negative)  Casos negativos correctamente predichos.
* FP (False Positive)  Casos negativos mal clasificados como positivos.
* FN (False Negative)  Casos positivos mal clasificados como negativos.

**Mtricas de Clasificacin (derivadas de la matriz)**
1. **Accuracy (Exactitud)**
Proporcin de predicciones correctas.
Puede ser engaosa si las clases estn desbalanceadas.

2. **Precision (Precisin o Valor Predictivo Positivo)**
De todas las veces que el modelo dijo positivo, cuntas fueron correctas?

3. **Recall (Sensibilidad o Tasa de Verdaderos Positivos)**
De todos los casos realmente positivos, cuntos detect el modelo?

4. **F1-Score**
Media armnica entre precisin y recall.
Muy til cuando hay desbalance de clases.

5. **Matriz extendida a mltiples clases**
Cuando hay ms de 2 clases, la matriz crece (ejemplo: 3x3, 4x4, etc.).

Las mtricas pueden calcularse en modo:
* Macro: promedio simple entre clases.
* Micro: promedio ponderado por nmero de muestras.
* Weighted: promedio ponderado que considera el desbalance de clases.

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Valores reales y predichos
y_real = [1, 0, 1, 1, 0, 1, 0]
y_pred = [1, 0, 1, 0, 0, 1, 1]

# Matriz de confusin
print("Matriz de Confusin:")
print(confusion_matrix(y_real, y_pred))

# Mtricas
print("Accuracy:", accuracy_score(y_real, y_pred))
print("Precision:", precision_score(y_real, y_pred))
print("Recall:", recall_score(y_real, y_pred))
print("F1 Score:", f1_score(y_real, y_pred))
```

### 7. Deteccin de outliers y valores atpicos

En este paso, identificaremos posibles outliers (valores atpicos) en las columnas numricas que podran influir en anlisis posteriores. Utilizaremos el Rango Intercuartlico (IQR) para definir los lmites y detectar outliers.

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers for numerical columns
numerical_cols = ['peso', 'altura', 'IMC']
for col in numerical_cols:
    outliers = detect_outliers_iqr(df1, col)
    print(f"\nOutliers detectados en la columna '{col}':")
    display(outliers)
```
