## Task
Modify the dataframe `df` to have 100 rows. Add a column named 'IMC' that calculates the Body Mass Index and a column named 'clasificacion' that classifies the individual based on their IMC (underweight, normal weight, overweight, etc.).

## Generate data

**Reasoning**:
Generate additional random data for name, weight, and height to reach 100 rows and concatenate it with the existing dataframe.



```python
import numpy as np

additional_rows_needed = 100 - len(df1)

# Generate random data for the new rows
random_names = [f"Person_{i}" for i in range(additional_rows_needed)]
random_weights = np.random.uniform(40, 100, additional_rows_needed) # realistic weight range
random_heights = np.random.uniform(1.50, 1.90, additional_rows_needed) # realistic height range

# Create a new DataFrame with the random data
new_data = {'nombre': random_names, 'peso': random_weights, 'altura': random_heights}
new_df = pd.DataFrame(new_data)

# Concatenate the original DataFrame with the new DataFrame
df1 = pd.concat([df1, new_df], ignore_index=True)

# Display the updated dataframe to verify the number of rows
display(df1)
```

## Calculate imc

### Subtask:
Add a new column named 'IMC' to the dataframe and calculate the Body Mass Index for each row using the 'peso' and 'altura' columns. The formula for BMI is weight (kg) / height (m)^2.


**Reasoning**:
Add a new column 'IMC' to the dataframe and calculate the BMI using the 'peso' and 'altura' columns.



```python
df1['IMC'] = df1['peso'] / (df1['altura'] ** 2)
display(df1.head())
```

## Classify based on imc

### Subtask:
Add a new column named 'clasificacion' and classify each individual based on their calculated IMC. The classifications will be based on standard BMI categories (e.g., Underweight, Normal weight, Overweight, Obesity).


**Reasoning**:
Define a function to classify IMC and apply it to the 'IMC' column to create the 'clasificacion' column.



```python
def classify_imc(imc):
    if imc < 18.5:
        return 'Underweight'
    elif 18.5 <= imc < 25:
        return 'Normal weight'
    elif 25 <= imc < 30:
        return 'Overweight'
    else:
        return 'Obesity'

df1['clasificacion'] = df1['IMC'].apply(classify_imc)
display(df1.head())
```

#EDA (ANÁLISIS DE DATOS EXPLORATORIO)

**Análisis de Datos Exploratorio (EDA, Exploratory Data Analysis)** es el **primer paso fundamental** cuando trabajas con cualquier dataset, ayuda a **conocer los datos, limpiarlos y entender sus patrones** antes de aplicar modelos más avanzados.

**PASOS**

1. **Cargar y revisar el dataset**

* Importar librerías (`pandas`, `numpy`, `matplotlib`, `seaborn`).
* Cargar el dataset con `pd.read_csv()` u otro método.
* Revisar las primeras filas con `df.head()`.
* Ver el tamaño de los datos: `df.shape`.

2. **Inspección general**

* Ver los **tipos de datos**: `df.info()`.
* Identificar **valores nulos o faltantes**: `df.isnull().sum()`.
* Revisar valores duplicados: `df.duplicated().sum()`.

3. **Estadística descriptiva**

* Para **variables numéricas**: `df.describe()`.
* Para **variables categóricas**: `df.describe(include="object")` o `value_counts()`.
* Medidas importantes: media, mediana, moda, rango, varianza, desviación estándar.

4. **Análisis univariado (una variable a la vez)**

* **Numéricas** → histogramas, boxplots (para ver distribución y outliers).
* **Categóricas** → tablas de frecuencia, gráficos de barras.

5. **Análisis bivariado (relaciones entre dos variables)**

* Correlaciones entre numéricas: `df.corr()`.
* Diagramas de dispersión (`scatterplot`) para relaciones.
* Tablas cruzadas (`pd.crosstab`) para categóricas.
* Gráficos combinados (ej: boxplot de variable numérica según categoría).

6. **Análisis multivariado**

* Mapas de calor de correlaciones (`heatmap`).
* Análisis de distribución conjunta (pairplot en `seaborn`).
* Clustering visual básico (si aplica).

7. **Detección de outliers y valores atípicos**

* Usar **boxplots** para detectar datos extremos.
* Comparar rangos intercuartílicos (IQR).

8. **Limpieza y preparación de los datos**

* Tratar valores nulos (eliminarlos o imputarlos).
* Normalizar o estandarizar variables numéricas.
* Codificar variables categóricas (OneHotEncoder, LabelEncoder).
* Eliminar duplicados o inconsistencias.

9. **Visualización y conclusiones**

* Graficar patrones clave (tendencias, distribuciones, correlaciones).
* Resumir hallazgos en un **informe corto**:

  * ¿Qué variables explican mejor el dataset?
  * ¿Qué problemas de calidad de datos existen?
  * ¿Qué hipótesis iniciales se pueden plantear?

📌 **En resumen**:
El EDA es un proceso iterativo que combina **estadística descriptiva + visualización de datos** para conocer la estructura y patrones antes de aplicar modelos de machine learning o análisis avanzados.

#REGRESIÓN LINEAL

Cuando trabajas con un modelo de regresión, necesitas métricas que te permitan medir **qué tan bien predice los valores continuos en tu dataset**.

 **Métricas de Regresión**
1. **Error Absoluto Medio (MAE – Mean Absolute Error)**

* Promedio de las diferencias absolutas entre valores reales y predichos.
* Fácil de interpretar (mide en las mismas unidades de la variable).
* Cuanto más pequeño, mejor.

2. **Error Cuadrático Medio (MSE – Mean Squared Error)**

* Penaliza más los errores grandes porque eleva al cuadrado.
* Se usa mucho en optimización de modelos.
* Entre más pequeño, mejor.

3. **Raíz del Error Cuadrático Medio (RMSE – Root Mean Squared Error)**

* Es la raíz del MSE.
* Muy útil porque vuelve al mismo rango de la variable dependiente.
* Más interpretable que el MSE.

4. **Error Absoluto Medio Porcentual (MAPE – Mean Absolute Percentage Error)**

* Expresa el error en porcentaje.
* Intuitivo, pero puede ser problemático si hay valores reales cercanos a 0.

5. **Coeficiente de Determinación (R²)**

* Indica la proporción de la variabilidad de la variable dependiente explicada por el modelo.
* Valores cercanos a 1 → buen ajuste.
* Valores negativos → el modelo es peor que predecir con la media.

6. **R² Ajustado (Adjusted R²)**

* Variante del R² que ajusta según el número de variables en el modelo.
* Evita que añadir variables irrelevantes aumente artificialmente el R².

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Supongamos que tienes:
y_real = [3, 5, 7, 10]
y_pred = [2.5, 5.1, 6.8, 9.5]

# MAE
mae = mean_absolute_error(y_real, y_pred)

# MSE
mse = mean_squared_error(y_real, y_pred)

# RMSE
rmse = np.sqrt(mse)

# R²
r2 = r2_score(y_real, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R²:", r2)
```

##MATRIZ DE CORRELACIÓN

Una matriz de correlación es una herramienta estadística que muestra **cómo se relacionan (linealmente) entre sí las variables numéricas de un dataset.**

🔎 ¿Qué es la correlación?

La correlación mide el grado de asociación lineal entre dos variables.
Su valor está entre -1 y 1:

+1 → correlación positiva perfecta (si una sube, la otra también).
-1 → correlación negativa perfecta (si una sube, la otra baja).
0 → no hay correlación lineal.

📊 ¿Qué es la matriz de correlación?

Es una tabla cuadrada que contiene todos los **coeficientes de correlación entre pares de variables numéricas.**
La diagonal principal siempre es 1, porque cada variable está perfectamente correlacionada consigo misma.

Se usa para:
* Detectar relaciones entre variables.
* Ver redundancia (variables muy correlacionadas).
* Prevenir multicolinealidad en modelos de regresión.

#MATRIZ DE CONFUSIÓN

Es una tabla de doble entrada que se utiliza para **evaluar el rendimiento de un modelo de clasificación.**

Muestra cuántas predicciones fueron correctas y equivocadas, organizadas por clases.

📌 Ejemplo de una matriz de confusión binaria (dos clases: Positivo y Negativo):
|                   | Predicho Positivo       | Predicho Negativo       |
| ----------------- | ----------------------- | ----------------------- |
| **Real Positivo** | TP (Verdadero Positivo) | FN (Falso Negativo)     |
| **Real Negativo** | FP (Falso Positivo)     | TN (Verdadero Negativo) |

🔎 Significado:

* TP (True Positive) → Casos positivos correctamente predichos.
* TN (True Negative) → Casos negativos correctamente predichos.
* FP (False Positive) → Casos negativos mal clasificados como positivos.
* FN (False Negative) → Casos positivos mal clasificados como negativos.

**Métricas de Clasificación (derivadas de la matriz)**
1. **Accuracy (Exactitud)**
Proporción de predicciones correctas.
Puede ser engañosa si las clases están desbalanceadas.

2. **Precision (Precisión o Valor Predictivo Positivo)**
De todas las veces que el modelo dijo “positivo”, ¿cuántas fueron correctas?

3. **Recall (Sensibilidad o Tasa de Verdaderos Positivos)**
De todos los casos realmente positivos, ¿cuántos detectó el modelo?

4. **F1-Score**
Media armónica entre precisión y recall.
Muy útil cuando hay desbalance de clases.

5. **Matriz extendida a múltiples clases**
Cuando hay más de 2 clases, la matriz crece (ejemplo: 3x3, 4x4, etc.).

Las métricas pueden calcularse en modo:
* Macro: promedio simple entre clases.
* Micro: promedio ponderado por número de muestras.
* Weighted: promedio ponderado que considera el desbalance de clases.

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Valores reales y predichos
y_real = [1, 0, 1, 1, 0, 1, 0]
y_pred = [1, 0, 1, 0, 0, 1, 1]

# Matriz de confusión
print("Matriz de Confusión:")
print(confusion_matrix(y_real, y_pred))

# Métricas
print("Accuracy:", accuracy_score(y_real, y_pred))
print("Precision:", precision_score(y_real, y_pred))
print("Recall:", recall_score(y_real, y_pred))
print("F1 Score:", f1_score(y_real, y_pred))
```

### 7. Detección de outliers y valores atípicos

En este paso, identificaremos posibles outliers (valores atípicos) en las columnas numéricas que podrían influir en análisis posteriores. Utilizaremos el Rango Intercuartílico (IQR) para definir los límites y detectar outliers.

```python
# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers for numerical columns
numerical_cols = ['peso', 'altura', 'IMC']
for col in numerical_cols:
    outliers = detect_outliers_iqr(df1, col)
    print(f"\nOutliers detectados en la columna '{col}':")
    display(outliers)
```
