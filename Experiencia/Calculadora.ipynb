#CALCULADORA SENCILLA

```python
import math
print("MENU:")
print("1: +")
print("2: -")
print("3: *")
print("4: /")
print("5: **")  #potenciación
print("6: raiz") #Radicación ** (1/n)
print("7: log") #logaritmación
print("8: sin") #seno
print("9: cos") #coseno
print("10: tan") #tangente
print()

opcion=input("Ingrese la opción deseada del menú: ") # la opción se guarda como cadena

if opcion== '1' or opcion=='+' or opcion== '2' or opcion=='-' or opcion=='*' or opcion== '3' or opcion== '4' or opcion=='/' or opcion== '5' or opcion=='**':
  numero_uno=float(input('Ingrese el primer número de la operación:'))# guarda como flotante (número real)
  numero_dos=float(input('Ingrese el segundo número de la operación:'))# guarda como flotante (número real)

if opcion== '6'or opcion=='raiz':
  numero_uno=float(input('Ingrese el valor de la raíz (índice):'))# guarda como flotante (número real)
  numero_dos=float(input('Ingrese el valor del radicando:'))

if opcion== '7' or opcion=='log':
  numero_uno=float(input('Ingrese el logaritmo a calcular:'))
  numero_dos=float(input('Ingrese la base del logaritmo a calcular:'))

if opcion== '8' or opcion=='sin' or opcion== '9' or opcion=='cos' or opcion== '10'or opcion=='tan':
  angulo_grados=float(input('Ingrese el ángulo en grados:'))

#DEFINE LA OPERACIÓN

if opcion=='1'or opcion=='+':
  print("Vamos a sumar(+) los dos valores ingresados:\n")
  print(f"La suma es: {numero_uno} + {numero_dos} = {numero_uno+numero_dos}")

if opcion=='2'or opcion=='-':
  print("Vamos a restar(-) los dos valores ingresados:\n")
  resta=numero_uno-numero_dos
  print(f"La resta es: {numero_uno} + {numero_dos} = {resta}")

if opcion=='3'or opcion=='*':
  print("Vamos a multiplicar(*) los dos valores ingresados:\n")
  mult=numero_uno*numero_dos
  print(f'La multiplicación es: {numero_uno} * {numero_dos} = {mult}')  #Tome los numeros que están dentro de las llaves y se muestre la multiplicación x*y=z

if opcion=='4'or opcion=='/':
  print("Vamos a dividir(/) los dos valores ingresados:\n")
  #VALIDANDO EL DENOMINADOR (debe ser diferente de cero)
  if numero_dos!=0:
    print(f'La división es: {numero_uno} / {numero_dos} = {numero_uno/numero_dos}')
  else:
    print("¡No es posible dividir entre cero!")

if opcion=='5'or opcion=='**':
  print("Vamos a realizar potenciación(**) de los dos valores ingresados:\n")
  print(f'La potenciación es: {numero_uno} ** {numero_dos} = {numero_uno**numero_dos}')

if opcion=='6'or opcion=='raiz':
  print("Vamos a calcular la raiz:\n")
  Raiz=numero_dos**(1/numero_uno)
  #print('El resultado es:',Raiz)
  print(f'La raíz {numero_uno} de {numero_dos} = {Raiz}')

if opcion=='7'or opcion=='log':
  print("Vamos a calcular el logaritmo:\n")
  resultado=math.log(numero_uno,numero_dos)
  #print("Resultado=",resultado_1)
  print(f'El logaritmo base {numero_dos} de {numero_uno} = {resultado}')

if opcion=='8'or opcion=='sin':
  print("Vamos a calcular seno del ángulo:\n")
  resultado=math.sin(math.radians(angulo_grados))  #CONVIERTO DE GRADO QUE INGRESA USUARIO A RADIANES Y LE CALCULO EL SENO
  print(f"El seno de {angulo_grados}° es :{resultado}")

if opcion=='9'or opcion=='cos':
  print("Vamos a calcular coseno del ángulo:\n")
  resultado=math.cos(math.radians(angulo_grados))  #CONVIERTO DE GRADO QUE INGRESA USUARIO A RADIANES Y LE CALCULO COSENO
  print(f"El coseno de {angulo_grados}° es :{resultado}")

if opcion=='10'or opcion=='tan':
 print("Vamos a calcular tangente del ángulo:\n")
 resultado=math.tan(math.radians(angulo_grados))
 print(f"La tangente de {angulo_grados}° es :{resultado}")
```

#CALCULADORA PLUS

```python
import math

# =========================
# FUNCIONES DE INTERFAZ
# =========================
def mostrar_titulo():
    print("\n=== CALCULADORA PLUS ===\n")

def mostrar_menu():
    print("MENU:\n")
    print("1: + (Suma)")
    print("2: - (Resta)")
    print("3: * (Multiplicación)")
    print("4: / (División)")
    print("5: ** (Potenciación)")
    print("6: raiz (Radicación)")
    print("7: log (Logaritmo base n)")
    print("8: sin (Seno)")
    print("9: cos (Coseno)")
    print("10: tan (Tangente)")
    print("11: Salir")
    print()

# =========================
# FUNCIONES DE ENTRADA
# =========================
def pedir_float(mensaje):   #Asegura que usuario siempre ingrese número válido
    while True:             #Se repite constantemente hasta que usuario ingrese número válido
        try:                # Código que podría generar una excepción. Si no ocurre, el código se ejecuta y se salta except.
            return float(input(mensaje))    #El return rompe el bucle infinito
        except ValueError:  #código que se ejecuta si ocurre esa excepción específica, en este caso número inválido
            print("⚠ Entrada inválida. Por favor, ingrese un número.")

def ingresar_datos():
    opcion = input("Ingrese la opción deseada del menú: ")

    if opcion in ['1', '+', '2', '-', '3', '*', '4', '/', '5', '**']:
        numero_uno = pedir_float('Ingrese el primer número: ')
        numero_dos = pedir_float('Ingrese el segundo número: ')
        return opcion, numero_uno, numero_dos, None

    elif opcion in ['6', 'raiz']:
        numero_uno = pedir_float('Ingrese el valor de la raíz (índice): ')
        numero_dos = pedir_float('Ingrese el radicando: ')
        return opcion, numero_uno, numero_dos, None

    elif opcion in ['7', 'log']:
        numero_uno = pedir_float('Ingrese el número al que desea calcular logaritmo: ')
        numero_dos = pedir_float('Ingrese la base del logaritmo: ')
        return opcion, numero_uno, numero_dos, None

    elif opcion in ['8', 'sin', '9', 'cos', '10', 'tan']:
        angulo_grados = pedir_float('Ingrese el ángulo en grados: ')
        return opcion, None, None, angulo_grados

    elif opcion in ['11', 'salir']:          #Da la opción de salir
        return opcion, None, None, None

    else:
        print("⚠ Opción no válida. Intente de nuevo.")
        return None, None, None, None

# =========================
# FUNCIONES DE CÁLCULO
# =========================
def formato(num):   #para redondear numeros a 4 decimales
    if isinstance(num, str):    #verifica si un objeto es una instancia de una clase específica o de sus subclases
        return num              # si es un mensaje de error, no formatear
    return f"{num:.4f}"

def calcular_suma(a, b): return a + b
def calcular_resta(a, b): return a - b
def calcular_multiplicacion(a, b): return a * b
def calcular_division(a, b):
    if b == 0:
        return "Error: División por cero"
    return a / b
def calcular_potenciacion(a, b): return a ** b
def calcular_radicacion(indice, radicando):
    if indice == 0:
        return "Error: El índice de la raíz no puede ser cero"
    try:                    #REVISARRRRR
        return radicando ** (1 / indice)
    except ValueError:
        return "Error: No se puede calcular raíz de un número negativo con índice par"
def calcular_logaritmo_base_n(valor, base):
    if valor <= 0 or base <= 0 or base == 1:
        return "Error: Logaritmo inválido. El número y la base deben ser > 0 y la base ≠ 1"
    return math.log(valor, base)
def calcular_seno(grados): return math.sin(math.radians(grados))
def calcular_coseno(grados): return math.cos(math.radians(grados))
def calcular_tan(grados): return math.tan(math.radians(grados))

# =========================
# PROCESAR Y MOSTRAR
# =========================
def procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos, angulo_grados):
    if opcion in ['1', '+']:
        print(f"{numero_uno} + {numero_dos} = {formato(calcular_suma(numero_uno, numero_dos))}")

    elif opcion in ['2', '-']:
        print(f"{numero_uno} - {numero_dos} = {formato(calcular_resta(numero_uno, numero_dos))}")

    elif opcion in ['3', '*']:
        print(f"{numero_uno} * {numero_dos} = {formato(calcular_multiplicacion(numero_uno, numero_dos))}")

    elif opcion in ['4', '/']:
        print(f"{numero_uno} / {numero_dos} = {formato(calcular_division(numero_uno, numero_dos))}")

    elif opcion in ['5', '**']:
        print(f"{numero_uno} ** {numero_dos} = {formato(calcular_potenciacion(numero_uno, numero_dos))}")

    elif opcion in ['6', 'raiz']:
        print(f"Raíz {numero_uno} de {numero_dos} = {formato(calcular_radicacion(numero_uno, numero_dos))}")

    elif opcion in ['7', 'log']:
        print(f"log base {numero_dos} de {numero_uno} = {formato(calcular_logaritmo_base_n(numero_uno, numero_dos))}")

    elif opcion in ['8', 'sin']:
        print(f"sin({angulo_grados}°) = {formato(calcular_seno(angulo_grados))}")

    elif opcion in ['9', 'cos']:
        print(f"cos({angulo_grados}°) = {formato(calcular_coseno(angulo_grados))}")

    elif opcion in ['10', 'tan']:
        print(f"tan({angulo_grados}°) = {formato(calcular_tan(angulo_grados))}")

# =========================
# PROGRAMA PRINCIPAL CON BUCLE
# =========================
def llamar_funciones():
    while True:
        mostrar_titulo()
        mostrar_menu()
        opcion, numero_uno, numero_dos, angulo_grados = ingresar_datos()

        if opcion in ['11', 'salir']:
            print("\n✅ Gracias por usar la calculadora. ¡Hasta luego!")
            break

        if opcion:
            procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos, angulo_grados)

        input("\nPresiona ENTER para continuar...")

# =========================
# EJECUTAR
# =========================
llamar_funciones()
```

#CALCULADORA PLUS CON MATCH-CASE

```python
import math

# =========================
# FUNCIONES DE INTERFAZ
# =========================
def mostrar_titulo():
    print("\n=== CALCULADORA PLUS ===\n")

def mostrar_menu():
    print("MENU:\n")
    print("1: + (Suma)")
    print("2: - (Resta)")
    print("3: * (Multiplicación)")
    print("4: / (División)")
    print("5: ** (Potenciación)")
    print("6: raiz (Radicación)")
    print("7: log (Logaritmo base n)")
    print("8: sin (Seno)")
    print("9: cos (Coseno)")
    print("10: tan (Tangente)")
    print("0: Salir")
    print()

# =========================
# FUNCIONES DE ENTRADA
# =========================

def ingresar_datos():
    opcion = input("Ingrese la opción deseada del menú: ")
    match opcion:
      case '1'|'+'|'2'| '-'|'3'|'*'|'4'|'/'|'5'|'**':
          numero_uno = pedir_float('Ingrese el primer número: ')
          numero_dos = pedir_float('Ingrese el segundo número: ')
          return opcion, numero_uno, numero_dos, None

      case '6'| 'raiz':
          numero_uno = pedir_float('Ingrese el valor de la raíz (índice): ')
          numero_dos = pedir_float('Ingrese el radicando: ')
          return opcion, numero_uno, numero_dos, None

      case '7'|'log':
          numero_uno = pedir_float('Ingrese el número al que desea calcular logaritmo: ')
          numero_dos = pedir_float('Ingrese la base del logaritmo: ')
          return opcion, numero_uno, numero_dos, None

      case '8'| 'sin'| '9'| 'cos'| '10'| 'tan':
          angulo_grados = pedir_float('Ingrese el ángulo en grados: ')
          return opcion, None, None, angulo_grados

      case '0'| 'salir':
          return opcion, None, None, None

      case _:
          print("⚠ Opción no válida. Intente de nuevo.")
          return None, None, None, None

# =========================
# FUNCIONES DE CÁLCULO
# =========================
def formato(num):   #para redondear numeros a 4 decimales
    if isinstance(num, str):    #verifica si un objeto es una instancia de una clase específica o de sus subclases
        return num  # si es un mensaje de error, no formatear
    return f"{num:.4f}"

def calcular_suma(a, b): return a + b
def calcular_resta(a, b): return a - b
def calcular_multiplicacion(a, b): return a * b
def calcular_division(a, b):
    if b == 0:
        return "Error: División por cero"
    return a / b
def calcular_potenciacion(a, b): return a ** b
def calcular_radicacion(indice, radicando):
    if indice == 0:
        return "Error: El índice de la raíz no puede ser cero"
    try:                    #REVISARRRRR
        return radicando ** (1 / indice)
    except ValueError:
        return "Error: No se puede calcular raíz de un número negativo con índice par"
def calcular_logaritmo_base_n(valor, base):
    if valor <= 0 or base <= 0 or base == 1:
        return "Error: Logaritmo inválido. El número y la base deben ser > 0 y la base ≠ 1"
    return math.log(valor, base)
def calcular_seno(grados): return math.sin(math.radians(grados))
def calcular_coseno(grados): return math.cos(math.radians(grados))
def calcular_tan(grados): return math.tan(math.radians(grados))

# =========================
# PROCESAR Y MOSTRAR
# =========================
def procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos, angulo_grados):
  match opcion:
    case '1'| '+':
        print(f"{numero_uno} + {numero_dos} = {formato(calcular_suma(numero_uno, numero_dos))}")

    case '2'| '-':
        print(f"{numero_uno} - {numero_dos} = {formato(calcular_resta(numero_uno, numero_dos))}")

    case '3'| '*':
        print(f"{numero_uno} * {numero_dos} = {formato(calcular_multiplicacion(numero_uno, numero_dos))}")

    case '4'| '/':
        print(f"{numero_uno} / {numero_dos} = {formato(calcular_division(numero_uno, numero_dos))}")

    case '5'| '**':
        print(f"{numero_uno} ** {numero_dos} = {formato(calcular_potenciacion(numero_uno, numero_dos))}")

    case '6'|'raiz':
        print(f"Raíz {numero_uno} de {numero_dos} = {formato(calcular_radicacion(numero_uno, numero_dos))}")

    case '7'|'log':
        print(f"log base {numero_dos} de {numero_uno} = {formato(calcular_logaritmo_base_n(numero_uno, numero_dos))}")

    case '8'|'sin':
        print(f"sin({angulo_grados}°) = {formato(calcular_seno(angulo_grados))}")

    case '9'| 'cos':
        print(f"cos({angulo_grados}°) = {formato(calcular_coseno(angulo_grados))}")

    case '10'|'tan':
        print(f"tan({angulo_grados}°) = {formato(calcular_tan(angulo_grados))}")

# =========================
# PROGRAMA PRINCIPAL CON BUCLE
# =========================
def llamar_funciones():
    while True:
        mostrar_titulo()
        mostrar_menu()
        opcion, numero_uno, numero_dos, angulo_grados = ingresar_datos()

        if opcion in ['0', 'salir']:
            print("\n✅ Gracias por usar la calculadora. ¡Hasta luego!")
            break

        if opcion:
            procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos, angulo_grados)

        input("\nPresiona ENTER para continuar...")

# =========================
# EJECUTAR
# =========================
llamar_funciones()
```
