#CICLO FOR

```python
frutas=["manzana","pera","piña","sandia"]
for fruta in frutas:
  print(fruta)
```

##Break

```python
ciudades=["Cali", "Bogota", "Medellin", "Barranquilla", "Cartagena"]
for ciudad in ciudades:
  print (ciudad)
  if ciudad =="Barranquilla":
    break

#Se detiene el ciclo cuando variable ciudad es Barranquilla
```

```python
ciudades=["Cali", "Bogota", "Medellin", "Barranquilla", "Cartagena"]
for ciudad in ciudades:
  if ciudad =="Barranquilla": #Si ciudad es igual a barranquilla
    break                     #sal del ciclo
  print (ciudad)              #sino imprime ciudad

#Se detiene el ciclo antes de llegar a Barranquilla
```

##Continue

```python
ciudades=["Cali", "Bogota", "Medellin", "Barranquilla", "Cartagena"]
for ciudad in ciudades:
  if ciudad =="Medellin":
    continue
  print (ciudad)

#Se salta la variable ciudad cuando están en Medellín y continua con las demás
```

##Función Range()

```python
'''
Recorrer un conjunto de código un número específico de veces
La función range() devuelve una secuencia de números que comienza en 0 por defecto, se incrementa en 1 (por defecto) y termina en un número especificado.
'''
for x in range(6):
  print(x)
```

```python
for x in range(2, 6):
  print(x)
#Especifica el valor desde que inicia
```

```python
for x in range(2, 30, 3):
  print(x)

#El último número especifica de cuánto es el incremento
```

##Uso de Else

```python
ciudades=["Cali", "Bogota", "Medellin", "Barranquilla", "Cartagena"]
for ciudad in ciudades:
  print (ciudad)
else:
  print("\nSe termino la lista")
#Especifica qué se debe realizar al finalizar el ciclo
#IMPORTANTE: NO se ejecuta si el ciclo es detenido por break
```

##Bucles anidados

```python
ciudades=["Cali", "Bogota", "Medellin"]
departamentos=["Valle","Cundinamarca","Antioquia"]
for ciudad in ciudades:
  for departamento in departamentos:
    print(ciudad,departamento)
#El ciclo interno se ejecuta para cada interacción del ciclo externo
```

##Pass

```python
'''
Los bucles for no pueden estar vacíos, pero si hay uno,
se coloca la declaración pass para evitar obtener un error (No se ejecuta).
'''
for x in [0, 1, 2]:
  pass
```

#CICLO WHILE

```python
variable =0
while variable < 10:
  print(variable)
  variable= variable + 1
  #variable += 1
```

```python
variable = 5
while variable >= 1:
  print(variable)
  variable= variable - 1
  #variable -= 1
```

##Break

```python
num = 3
while num < 9:
  print(num)
  if num == 7:
    break
  num += 2
#Se detiene el ciclo aún si la condición del while es verdadera
```

##Continue

```python
num = 3
while num < 9:
  num += 2
  if num == 7:
    continue
  print(num)

#Se salta el número que cumpla la condición if y se continua con el siguiente
```

##Else

```python
num = 3
while num < 12:
  print(num)
  num += 2
else:
  print("\nSe termino la ejecución")

#Se ejecuta el bloque de else una vez la condición de while no es verdadera
```

##Otros

```python
# Lista de números
numeros = [2, 4, 6, 8, 10]

i = 0
suma = 0

while i < len(numeros): #Mientras posición de la lista sea menor a la longitud de la lista
    suma += numeros[i]  #a variable suma, actualizala sumando el número que está en la posición de i
    i += 1              #Actualiza la variable i, para pasar al siguiente número

print("La suma de los números es:", suma)

#Recorre la lista sumando cara número
```

```python
# Lista de nombres
nombres = ["Ana", "Luis", "Carlos", "María", "Sofía"]

buscado = "Luisa"
i = 0
encontrado = False

while i < len(nombres): #Mientras posición de la lista sea menor a la longitud de la lista
    if nombres[i] == buscado: #Si, nombres en la posición i, es igual a la variable buscado
        encontrado = True     #variable encontrado pasa a ser verdadera
        break
    i += 1

if encontrado:
    print("El nombre", buscado, "sí está en la lista.")
else:
    print("El nombre", buscado, "NO está en la lista.")
```

```python
# Lista de números
numeros = [5, 3, 8, 1, 4]

while numeros:  # Mientras la lista no esté vacía
    elemento = numeros.pop()  # elimina el último elemento
    print("Sacando:", elemento, " -> Lista ahora:", numeros)

#Cada iteración elimina el último número (.pop()) y lo muestra.
```
