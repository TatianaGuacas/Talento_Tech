#Triqui

```python
# Juego de Triqui (Tres en lÃ­nea)

# Tablero inicial
tablero = [" " for _ in range(9)]

# FunciÃ³n para mostrar el tablero
def mostrar_tablero():
    print("\n")
    for i in range(0, 9, 3):
        print(tablero[i], "|", tablero[i+1], "|", tablero[i+2])
        if i < 6:
            print("--+---+--")
    print("\n")

# FunciÃ³n para verificar si alguien ganÃ³
def verificar_ganador(simbolo):
    # Combinaciones ganadoras
    combinaciones = [
        [0,1,2], [3,4,5], [6,7,8],  # filas
        [0,3,6], [1,4,7], [2,5,8],  # columnas
        [0,4,8], [2,4,6]            # diagonales
    ]
    for combo in combinaciones:
        if tablero[combo[0]] == tablero[combo[1]] == tablero[combo[2]] == simbolo:
            return True
    return False

# FunciÃ³n para verificar empate
def verificar_empate():
    return " " not in tablero

# Juego principal
jugador_actual = "X"

while True:
    mostrar_tablero()
    try:
        posicion = int(input(f"Jugador {jugador_actual}, elige posiciÃ³n (1-9): ")) - 1
        if posicion < 0 or posicion > 8 or tablero[posicion] != " ":
            print("PosiciÃ³n invÃ¡lida, intenta de nuevo.")
            continue
    except ValueError:
        print("Por favor ingresa un nÃºmero vÃ¡lido.")
        continue

    # Colocar la jugada
    tablero[posicion] = jugador_actual

    # Verificar si hay ganador
    if verificar_ganador(jugador_actual):
        mostrar_tablero()
        print(f"ğŸ‰ Â¡Jugador {jugador_actual} gana!")
        break

    # Verificar si hay empate
    if verificar_empate():
        mostrar_tablero()
        print("ğŸ¤ Â¡Es un empate!")
        break

    # Cambiar de jugador
    jugador_actual = "O" if jugador_actual == "X" else "X"
```

#Ahorcado

```python
#Ahorcado de animales

import random

#Lista de palabras
animales= ["jirafa", "elefante", "rinoceronte", "serpiente", "ballena", "flamenco", "cocodrilo"]

#Elegir una palabra aleatoria
animal= random.choice(animales)
letra_adivinada = []
vida = 5

print("ğŸ‰ Bienvenido al juego del Ahorcado de animales ğŸ‰")

#Ciclo principal
while vida >0:
  progreso = "" #Se muestra el progreso de la palabra
  for letra in animal:
    if letra in letra_adivinada:
      progreso += letra + " "

    else:
      progreso += "_ "

  print ("\nPalabra:", progreso)

  #Verifica si gana adivinando todas las letras
  if "_" not in progreso:
    print("ğŸ‰ Â¡Felicidades! Adivinaste la palabra:", animal)
    break

  #Pedir intento (letra o palabra)
  intento = input("ğŸ‘‰ Ingresa una letra o adivina la palabra completa: ").lower() #.lower devuelve en minusculas

  #Si intento es la palabra completa
  if intento == animal:
    print("ğŸ‰ Â¡IncreÃ­ble! Adivinaste la palabra completa:", animal)
    break

  #Validar entrada
  if len(intento) == 1 and intento.isalpha():   #isalpha verifica que caracter sea string y cadena no estÃ© vacÃ­a. Devuelve true
    if intento in letra_adivinada:              #revisa si la letra ya se usÃ³
      print("Ya intentaste esa letra, prueba otra.")
      continue

    #agrega letra a la lista de intentos
    letra_adivinada.append(intento)

    #comprueba si letra estÃ¡ en la palabra
    if intento in animal:
      print("âœ… Â¡Bien! La letra estÃ¡ en la palabra.")

    else:
      vida -= 1
      print(f"âŒ La letra no estÃ¡. Te quedan {vida} vidas.")

  else:
    print("âš ï¸ Ingresa solo una letra vÃ¡lida o la palabra completa.")

#Si se queda sin vidas
if vida == 0:
  print(f"ğŸ’€ Perdiste! La palabra era: {animal}")
```
