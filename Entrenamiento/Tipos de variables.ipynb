##Operación en listas

```python
frutas=["pera","manzana","naranja"]
print(frutas)
```

```python
#Agregar variable
frutas.append("Mango")
print(frutas)
```

```python
#Insertar en una posición
frutas.insert(2,"Limon")
print(frutas)
```

```python
#Eliminar posición        NO FUNCIONA SI LA PALABRA TIENE TILDE?
frutas.remove('Limon')
print(frutas)
```

```python
#Retorne elementos que están en rango
frutas[2:5]   #[:5] trae desde posición 0
```

```python
#Sumar variables de la lista
numeros=[8, 5, 1]
suma=numeros[0]+numeros[1]+numeros[2]
print(suma)
```

```python
suma=sum(numeros)
print(suma)
```

```python
#Longitud de la lista
print(len(numeros))

#Máximo
print(max(numeros))

#Mínimo
print(min(numeros))

#Posicion de un elemento
print(frutas.index("Mango"))
```

##Tipo diccionarios

```python
Persona={
    "nombre":"Tatiana",
"apellido":"Guacas",
"edad":23
}

print(Persona.get("apellido"))
#print(Persona("apellido"))
```

```python
#Eliminar clave
del Persona["edad"]
print(Persona)
```

```python
Persona.pop("apellido")   #muestra lo que se va a eliminar
```

```python
#Diccionario anidado
alumnos={
    "alumno1":{"nombre":"Tatiana", "nota":[5, 3, 4]},
    "alumno2":{"nombre":"Luis", "nota":[4.5, 5, 4.5]},
    "alumno3":{"nombre":"Lina", "nota":[4, 3, 5]}
}

print(alumnos["alumno1"])

#imprimir lo que contiene nota de alumno2
print(alumnos["alumno2"]["nota"])

#imprimir lo que contiene nota de alumno3 en posición 2
print(alumnos["alumno3"]["nota"][2])
```

```python
#Anadir al diccionario
alumnos["alumno4"]="Hola"
print(alumnos)
```

```python
#Saber cuáles claves tiene
print(alumnos.keys())
```

```python
#Saber los valores de las claves
print(alumnos.values())
```

```python
#Saber los items
print(alumnos.items())
```

```python
#lo que tenga d1 actualice a d2
d1={'A':1, 'B':2}
d2={'A':0, 'C':200}
d1.update(d2)
print(d1)
```
